```python
import tushare as ts
import matplotlib.pyplot as plt
import mplfinance as mpf

# 初始化pro接口
pro = ts.pro_api('c8ffaec9e6ade215441734a05c3c4ddd4ecdf6cd98c108ec182b0e4e')
```

```python
# 拉取数据
df = pro.daily(**{
    "ts_code": "002176.SZ",
    "trade_date": "",
    "start_date": 20220701,
    "end_date": 20220807,
    "offset": "",
    "limit": ""
}, fields=[
    "ts_code",
    "trade_date",
    "open",
    "high",
    "low",
    "close",
    "pre_close",
    "change",
    "pct_chg",
    "vol",
    "amount"
])
print(df)
```

```python

from matplotlib.pylab import date2num
import datetime

def date_to_num(dates):
    num_time = []
    for date in dates:
       date_time = datetime.datetime.strptime(date, "%Y%m%d")
       num_date = date2num(date_time)
       num_time.append(num_date)
    return num_time

df_arr = df.values
df_arr[:,0] = date_to_num(df)

print(date_to_num(df["trade_date"]))
```
